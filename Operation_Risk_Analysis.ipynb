# AGING ANALYSIS (How long cases are pending)

```python
today = pd.Timestamp.today()

open_cases['pending_days'] = (today - open_cases['Created_Date']).dt.days

open_cases['pending_days'].describe()
```

# CRITICAL + HIGH PRIORITY OLD CASES (ESCALATION LIST)

```python
# Now we identify dangerous open cases that need immediate action.
critical_old_cases = open_cases[
    (open_cases['Priority'].isin(['High', 'Critical'])) &
    (open_cases['pending_days'] > 365)
]

critical_old_cases.shape
```
